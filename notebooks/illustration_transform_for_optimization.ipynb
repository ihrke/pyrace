# Illustration of the transformation used in pyrace

The optimizer is operating on [-inf,+inf] but some parameters can only take on values in certain intervals, e.g. [0,1] in case of probabilities.

A special function mapping from [a,b] -> [-inf, +inf] -> [a,b] is used by the optimizers internally.

This is the definiton of the function:

```python
import numpy as np
import pylab as pl
from pyrace import trans_logistic as trans
```

## if boundaries are unrestricted, no transformation is applied:

```python
a=-np.infty
b=np.infty
x=np.linspace(-1000,1000,10000)
y=trans(x, a, b)
x2=trans(y,a,b, inverse=True)

pl.figure(figsize=(10,5))
pl.subplot(1,3,1)
pl.plot(x,y)
pl.title('[{a},{b}]->[-inf,+inf]'.format(a=a,b=b))
pl.subplot(1,3,2)
pl.plot(y,x2)
pl.title('[-inf,+inf]->[{a},{b}]'.format(a=a,b=b))
pl.subplot(1,3,3)
pl.plot(x,x2)
pl.title('[-inf,+inf]->[{a},{b}]->[-inf,+inf]'.format(a=a,b=b))
```

## if one of the two boundaries is infinite, a logarithmic transform is used

```python
a=2
b=np.infty
x=np.linspace(a,1000,10000)
y=trans(x, a, b)
x2=trans(y,a,b, inverse=True)

pl.figure(figsize=(10,5))
pl.subplot(1,3,1)
pl.plot(x,y)
pl.title('[{a},{b}]->[-inf,+inf]'.format(a=a,b=b))
pl.subplot(1,3,2)
pl.plot(y,x2)
pl.title('[-inf,+inf]->[{a},{b}]'.format(a=a,b=b))
pl.subplot(1,3,3)
pl.plot(x,x2)
pl.title('[-inf,+inf]->[{a},{b}]->[-inf,+inf]'.format(a=a,b=b))
```
