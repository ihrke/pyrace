# Specifying a Model (all for Stop-Signal Models)
 
To specify a pyrace model, you can either directly derive a subclass of an existing model or use the `pyrace.ModelTable` class which has a high-level interface.

```python
import pylab as pl
import pyrace as pr
from pyrace import ParMap # just convenient for model specification
```

## first we need to specify a design 
   - here we use a single manipulation (control vs. experimental) and stimuli are presented left or right

```python
factors=[{'manipulation':['control', 'exp']},
         {'stimulus':['left', 'right']}]
responses=['left', 'right']
design=pr.Design(factors, responses, 'stimulus', name='singlego')
print design
```

## select a class of models

* you can choose any model which uses the same accumulator-type for all accumulators
* currently, the following ones are defined

```python
print [mod.__name__ for mod in pr.stop_signal_models]
```

## Specify dependencies for all parameters

* here I fix variance of drift-rate (sv), boundary (b) and probability of go-failures and trigger failures (pgf/ptf)
* ter and A are constant across all conditions
* drift rate varies between correct and incorrect responses and stop-drift rate is the same as incorrect response drift

The table-based representation shown here specifies which model-parameter maps to which accumulator in which condition and given which response or GO/STOP trials.

```python
modtab=pr.ModelTable('testModel1', design, pr.pSSLBA,
                     fixed={'sv':1,'b':1, 'pgf':0, 'ptf':0},
                     ter=ParMap('ter'),
                     A=ParMap('A'),
                     V=ParMap('v', correct=True),
                     v=ParMap('v', correct=False)
                     )
display(modtab)
```

if you think this model does what you want, you can look at the code of an implementation of that model.

You can see, that a parameter specification along with a mapping to each accumulator is produced.

```python
print modtab.generate_model_str()
```

more conveniently, you can get a class instance directly, that allows you to derive model instances

```python
model_class=modtab.generate_model_class()
mod=model_class( (0.1, 1, 0.1, .5) )
print mod

pl.figure(figsize=(10,10))
mod.plot_model(lims=(0,3))
```

## More complicated models now

using the same design. 

Note, you can give any parameters to `ParMap` and they will be interpreted as indexing into the model-table (the HTML-table above).

If you miss fields, you will be warned.

```python
## e.g. here I have differnt correct-drifts for experimental manipulation vs. control
modtab=pr.ModelTable('testModel1b', design, pr.pSSLBA,
                     fixed={'sv':1,'b':1, 'pgf':0, 'ptf':0},
                     ter=ParMap('ter'),
                     A=ParMap('A'),
                     Vc=ParMap('v', manipulation='control', correct=True, gostop='go'),
                     Ve=ParMap('v', manipulation='exp', correct=True, gostop='go'),
                     v=ParMap('v', correct=False, gostop='go'),
                     Vs=ParMap('v', gostop="stop")
                     )
print modtab
```

```python
## e.g. here I forget to specify a drift rate for the stop-trials
modtab=pr.ModelTable('testModel2', design, pr.pSSLBA,
                     fixed={'sv':1,'b':1, 'pgf':0, 'ptf':0},
                     ter=ParMap('ter'),
                     A=ParMap('A'),
                     V=ParMap('v', correct=True, gostop='go'),
                     v=ParMap('v', correct=False, gostop='go')
                     )
```

```python
## rectified here
modtab=pr.ModelTable('testModel3', design, pr.pSSLBA,
                     fixed={'sv':1,'b':1, 'pgf':0, 'ptf':0},
                     ter=ParMap('ter'),
                     A=ParMap('A'),
                     V=ParMap('v', correct=True, gostop='go'),
                     v=ParMap('v', correct=False, gostop='go'),
                     Vs=ParMap('v', gostop='stop')
                     )
modtab
```

### you can also give mappings from model to accumulator parameters

(note that in the generated code, b=A+B in the accumulator definitions)

```python
## here I define a model that reparametrizes the boundary b=A+B 
## (i.e., the model optimizes B and projects internally to b using A+B)
modtab=pr.ModelTable('testModel4', design, pr.pSSLBA,
                     fixed={'sv':1, 'pgf':0, 'ptf':0},
                     ter=ParMap('ter'),
                     A=ParMap('A'),
                     B=ParMap('b', mapping="A+B"),
                     V=ParMap('v', correct=True, gostop='go'),
                     v=ParMap('v', correct=False, gostop='go'),
                     Vs=ParMap('v', gostop='stop')
                     )
print modtab.generate_model_str()
```

* for optimization you should specify reasonable bounds
* if no bounds are specified, the maximum possible ones defined by each accumulator are used
   - however, these can be sometimes wrong, because of non-trival mappings

```python
## here I define a model that reparametrizes the boundary b=A+B 
## (i.e., the model optimizes B and projects internally to b using A+B)
modtab=pr.ModelTable('testModel4', design, pr.pSSLBA,
                     fixed={'sv':1, 'pgf':0, 'ptf':0},
                     ter=ParMap('ter', bounds=(0,5)),
                     A=ParMap('A', bounds=(0,5)),
                     B=ParMap('b', mapping="A+B", bounds=(0,5)),
                     V=ParMap('v', correct=True, gostop='go', bounds=(-5,5)),
                     v=ParMap('v', correct=False, gostop='go', bounds=(-5,5)),
                     Vs=ParMap('v', gostop='stop', bounds=(-5,5))
                     )
modcl=modtab.generate_model_class()
mod=modcl()
print mod
mod.plot_model()
```
